```python
%matplotlib inline
# Some imports
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a synthetic dataset
np.random.seed(42)
n = 500

data = pd.DataFrame({
    "soil_moisture": np.random.uniform(10, 60, n),
    "temperature": np.random.uniform(15, 40, n),
    "rainfall": np.random.uniform(0, 20, n),
})

# synthetic yield function
data["yield"] = (
    0.3 * data["soil_moisture"]
    - 0.2 * data["temperature"]
    + 0.5 * data["rainfall"]
    + np.random.normal(0, 2, n)
)
```

```python
# Plot the synthetic dataset
sns.set(style="whitegrid")

plt.figure(figsize=(6,4))
sns.scatterplot(
    data=data,
    x="soil_moisture",
    y="yield",
    alpha=0.6
)
plt.title("Soil moisture vs crop yield (synthetic data)")
plt.xlabel("Soil moisture (%)")
plt.ylabel("Yield (arbitrary units)")
plt.tight_layout()
plt.show()
```

```python
# Simple linear regression model
X = data[["soil_moisture", "temperature", "rainfall"]]
y = data["yield"]

model = LinearRegression()
model.fit(X, y)
```

```python
# Evaluate the model
pred = model.predict(X)
print("MAE:", mean_absolute_error(y, pred))
```

```python
sample = pd.DataFrame([{
    "soil_moisture": 35,
    "temperature": 28,
    "rainfall": 3
}])

predicted_yield = model.predict(sample)[0]
predicted_yield
```

```python
# Model fit evaluation plot
plt.figure(figsize=(6,4))
plt.scatter(y, pred, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], "r--")
plt.xlabel("True yield")
plt.ylabel("Predicted yield")
plt.title("Model prediction vs ground truth")
plt.tight_layout()
plt.show()
```

```python
coef_df = pd.DataFrame({
    "feature": X.columns,
    "coefficient": model.coef_
})

plt.figure(figsize=(6,4))
sns.barplot(data=coef_df, x="feature", y="coefficient")
plt.title("Feature influence on yield prediction")
plt.tight_layout()
plt.show()
```
